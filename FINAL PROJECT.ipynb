## <span style="color: blue;">MODEL DATA</span>

```python
X = data.iloc[:, :-1]
Y = data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.7)
x_train
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly=PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_train_poly.shape
```

```python
x_train_poly
```

```python
model=LinearRegression()
model.fit(x_train_poly,y_train)
a=model.coef_
a
```

```python
y_train_pred=model.predict(x_train_poly)
```

```python
x_test_poly = poly.fit_transform(x_test)
x_test_poly.shape
```

```python
y_test_pred=model.predict(x_test_poly)
```

## <span style="color: blue;">TRAIN AND TEST MODELS</span>

```python
plt.scatter(x_train["Cement"],y_train)
plt.scatter(x_train["Cement"],y_train_pred)
plt.xlabel("conc")
plt.ylabel("strength")
plt.legend(["Actual","Predicted"])
```

```python
plt.scatter(x_test["Cement"],y_test)
plt.scatter(x_test["Cement"],y_test_pred)
plt.xlabel("conc")
plt.ylabel("strength")
plt.legend(["Actual","Predicted"])
```

## <span style="color: blue;">Polynomial (Quantitive measure)</span>



```python
from sklearn import metrics
trainMSE=metrics.mean_squared_error(y_train,y_train_pred)
trainRMSE=np.sqrt(trainMSE)
trainRMSE
```

```python
testMSE=metrics.mean_squared_error(y_test,y_test_pred)
testRMSE=np.sqrt(testMSE)
testRMSE
```

## <span style="color: blue;">Linear (Quantitive measure)</span>


```python
plt.scatter(y_train,y_train_pred,color='b')
plt.plot([y_train.min(), y_train.max()], [y_train_pred.min(), y_train_pred.max()], color = 'black', lw=2)
plt.xlabel("Y_train")
plt.ylabel("Y_train_pred")
plt.title("Predictions vs. actual values")
```

```python
plt.scatter(y_test,y_test_pred,color='b')
plt.plot([y_test.min(), y_test.max()], [y_test_pred.min(), y_test_pred.max()], color = 'black', lw=2)
plt.xlabel("Y_test")
plt.ylabel("Y_test_pred")
plt.title("Predictions vs. actual values")
```

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
y_train_pred_lin=linear_model.predict(x_train)
y_test_pred_lin=linear_model.predict(x_test)
```

```python
train_linRMSE=np.sqrt(metrics.mean_squared_error(y_train,y_train_pred_lin))
train_linRMSE
```

```python
test_linRMSE=np.sqrt(metrics.mean_squared_error(y_test,y_test_pred_lin))
test_linRMSE
```

## <span style="color: blue;">DECISION TREES</span>

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
dtr = DecisionTreeRegressor()

dtr.fit(x_train, y_train)

y_pred_dtr = dtr.predict(x_test)

print("Model\t\t\t\t\t RMSE \t\t MSE \t\tMAE \t\tR2")
print("""DecisionTreeRegressor \t\t {:.2f} \t\t{:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(y_test, y_pred_dtr)),mean_squared_error(y_test, y_pred_dtr),
            mean_absolute_error(y_test, y_pred_dtr), r2_score(y_test, y_pred_dtr)))

plt.scatter(y_test, y_pred_dtr, color='blue', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'black', lw=2)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Decision Tree Regressor')
plt.show()

```

```python
model_names = ['Linear Train RMSE', 'Linear Test RMSE', 'Polynomial Train RMSE', 'Polynomial Test RMSE', 'Decision Tree RMSE']
rmse_values = [train_linRMSE, test_linRMSE, trainRMSE, testRMSE, np.sqrt(mean_squared_error(y_test, y_pred_dtr))]

plt.figure(figsize=(14, 6))
bars = plt.bar(model_names, rmse_values, color=['blue', 'green', 'red', 'purple', 'orange'])

plt.xlabel('Model')
plt.ylabel('RMSE')
plt.title('RMSE Comparison of Different Models')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```
